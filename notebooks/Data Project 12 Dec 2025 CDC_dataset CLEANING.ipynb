```python
import pandas as pd
cdc = pd.read_csv("/data/CDCSocialVulnerabilityIndexFlorida_dataset.csv"
                  , dtype={"FIPS": "string"})
```

```python
cdc.head()
cdc.info()
cdc.describe()
```

```python
cdc.dropna()
cdc.drop_duplicates()
```

```python
# GEOID (tract key) 
cdc["GEOID"] = cdc["FIPS"].str.zfill(11)

# South Florida county FIPS (first 5 digits of GEOID) 
sf_county_fips = ["12011", "12086", "12099", "12087"]  # Broward, Miami-Dade, Palm Beach, Monroe

# Columns to keep (core + interpretable + flags) 
core_cols = [
    "GEOID", "ST", "STATE", "ST_ABBR", "COUNTY", "STCNTY", "LOCATION", "AREA_SQMI",
    "RPL_THEME1", "RPL_THEME2", "RPL_THEME3", "RPL_THEME4", "RPL_THEMES",
    "SPL_THEME1", "SPL_THEME2", "SPL_THEME3", "SPL_THEME4", "SPL_THEMES",
]

explain_cols = [
    "EP_POV150", "EP_UNEMP", "EP_HBURD", "EP_NOHSDP", "EP_UNINSUR",
    "EP_AGE65", "EP_AGE17", "EP_DISABL", "EP_SNGPNT", "EP_LIMENG",
    "EP_MINRTY", "EP_MUNIT", "EP_MOBILE", "EP_CROWD", "EP_NOVEH", "EP_GROUPQ",
]

flag_cols = ["F_THEME1", "F_THEME2", "F_THEME3", "F_THEME4", "F_TOTAL"]

keep = [c for c in (core_cols + explain_cols + flag_cols) if c in cdc.columns]
cdc_clean = cdc[keep].copy()
```

```python
# Convert numeric columns
num_cols = [c for c in cdc_clean.columns if c.startswith(("RPL_", "SPL_", "EP_"))]
cdc_clean[num_cols] = cdc_clean[num_cols].apply(pd.to_numeric, errors="coerce")

# Enforce valid ranges
# RPL_* are percentile ranks in [0, 1]
for c in [c for c in cdc_clean.columns if c.startswith("RPL_")]:
    cdc_clean.loc[(cdc_clean[c] < 0) | (cdc_clean[c] > 1), c] = pd.NA

# EP_* are percents in [0, 100]
for c in [c for c in cdc_clean.columns if c.startswith("EP_")]:
    cdc_clean.loc[(cdc_clean[c] < 0) | (cdc_clean[c] > 100), c] = pd.NA
```

```python
# Renames
rename_map = {
    "RPL_THEME1": "svi_theme1_pctile",
    "RPL_THEME2": "svi_theme2_pctile",
    "RPL_THEME3": "svi_theme3_pctile",
    "RPL_THEME4": "svi_theme4_pctile",
    "RPL_THEMES": "svi_overall_pctile",
    "SPL_THEME1": "svi_theme1_score",
    "SPL_THEME2": "svi_theme2_score",
    "SPL_THEME3": "svi_theme3_score",
    "SPL_THEME4": "svi_theme4_score",
    "SPL_THEMES": "svi_overall_score",
}
cdc_clean = cdc_clean.rename(columns=rename_map)
```

```python
# Sanity checks
print("Rows:", len(cdc_clean))
print("Duplicate GEOIDs:", int(cdc_clean["GEOID"].duplicated().sum()))
print("Missing overall pctile:", round(cdc_clean["svi_overall_pctile"].isna().mean() * 100, 2), "%")
```

```python
# Save (South Florida subset)
cdc_sf = cdc_clean[cdc_clean["GEOID"].str[:5].isin(sf_county_fips)].copy()
cdc_sf.to_parquet("/data/CDCSocialVulnerabilityIndexFlorida_dataset CLEANED.parquet", index=False)
cdc_sf.to_csv("/data/CDCSocialVulnerabilityIndexFlorida_dataset CLEANED.csv", index=False)
```
