```python
import pandas as pd
import re

nri = pd.read_csv(
    "/Users/acsoteldo/Desktop/datasets/Data Project 12 Dec 2025/data/FEMANationalRiskIndex_dataset.csv",
    dtype={
        "State FIPS Code": "string",
        "County FIPS Code": "string",
        "Census Tract FIPS Code": "string",
        "State-County FIPS Code": "string"
    },
    low_memory=False
)
```

```python
nri.head()
nri.info()
nri.describe()
```

```python
nri.dropna()
nri.drop_duplicates()
```

```python
# GEOID (11-digit tract FIPS)
# Prefer Census Tract FIPS Code if it already looks like 11 digits
if "Census Tract FIPS Code" in nri.columns:
    nri["GEOID"] = nri["Census Tract FIPS Code"].str.replace(r"\.0$", "", regex=True).str.zfill(11)
else:
    # Fallback build
    nri["GEOID"] = (
        nri["State FIPS Code"].str.zfill(2)
        + nri["County FIPS Code"].str.zfill(3)
        + nri["Census Tract"].astype(str).str.replace(r"\.0$", "", regex=True).str.zfill(6)
    )
```

```python
# Filter to Florida
nri_fl = nri[nri["State Name Abbreviation"].eq("FL")].copy()

# Filter to South Florida counties
# Broward=011, Miami-Dade=086, Palm Beach=099, Monroe=087 (state=12)
# County FIPS code is 3 digits; GEOID first 5 digits identify county
sf_county_geoid_prefix = ["12011", "12086", "12099", "12087"]
nri_sf = nri_fl[nri_fl["GEOID"].str[:5].isin(sf_county_geoid_prefix)].copy()
```

```python
# Keep only Composite + social vulnerability/resilience + flood/hurricane hazard metrics
keep = [
    "GEOID",
    "State Name", "State Name Abbreviation",
    "County Name", "County FIPS Code",
    "Population (2020)",
    "Area (sq mi)",

    # Composite risk + value/score/rating
    "National Risk Index - Value - Composite",
    "National Risk Index - Score - Composite",
    "National Risk Index - Rating - Composite",
    "Expected Annual Loss - Total - Composite",
    "Expected Annual Loss - Score - Composite",
    "Expected Annual Loss - Rating - Composite",

    # Social vulnerability & resilience
    "Social Vulnerability - Score",
    "Social Vulnerability - Rating",
    "Community Resilience - Score",
    "Community Resilience - Rating",

    # Flood + hurricane hazard metrics (EAL + risk index score/value/rating)
    "Coastal Flooding - Expected Annual Loss - Total",
    "Coastal Flooding - Expected Annual Loss Score",
    "Coastal Flooding - Hazard Type Risk Index Score",
    "Coastal Flooding - Hazard Type Risk Index Value",
    "Coastal Flooding - Hazard Type Risk Index Rating",

    "Inland Flooding - Expected Annual Loss - Total",
    "Inland Flooding - Expected Annual Loss Score",
    "Inland Flooding - Hazard Type Risk Index Score",
    "Inland Flooding - Hazard Type Risk Index Value",
    "Inland Flooding - Hazard Type Risk Index Rating",

    "Hurricane - Expected Annual Loss - Total",
    "Hurricane - Expected Annual Loss Score",
    "Hurricane - Hazard Type Risk Index Score",
    "Hurricane - Hazard Type Risk Index Value",
    "Hurricane - Hazard Type Risk Index Rating",
]

keep = [c for c in keep if c in nri_sf.columns]
nri_sf_clean = nri_sf[keep].copy()
```

```python
# Rename to analysis-friendly names
rename = {
    "Population (2020)": "pop_2020",
    "Area (sq mi)": "area_sqmi",

    "National Risk Index - Value - Composite": "nri_value",
    "National Risk Index - Score - Composite": "nri_score",
    "National Risk Index - Rating - Composite": "nri_rating",

    "Expected Annual Loss - Total - Composite": "eal_total",
    "Expected Annual Loss - Score - Composite": "eal_score",
    "Expected Annual Loss - Rating - Composite": "eal_rating",

    "Social Vulnerability - Score": "social_vuln_score",
    "Social Vulnerability - Rating": "social_vuln_rating",
    "Community Resilience - Score": "resilience_score",
    "Community Resilience - Rating": "resilience_rating",

    "Coastal Flooding - Expected Annual Loss - Total": "coastal_flood_eal_total",
    "Coastal Flooding - Expected Annual Loss Score": "coastal_flood_eal_score",
    "Coastal Flooding - Hazard Type Risk Index Score": "coastal_flood_risk_score",
    "Coastal Flooding - Hazard Type Risk Index Value": "coastal_flood_risk_value",
    "Coastal Flooding - Hazard Type Risk Index Rating": "coastal_flood_risk_rating",

    "Inland Flooding - Expected Annual Loss - Total": "inland_flood_eal_total",
    "Inland Flooding - Expected Annual Loss Score": "inland_flood_eal_score",
    "Inland Flooding - Hazard Type Risk Index Score": "inland_flood_risk_score",
    "Inland Flooding - Hazard Type Risk Index Value": "inland_flood_risk_value",
    "Inland Flooding - Hazard Type Risk Index Rating": "inland_flood_risk_rating",

    "Hurricane - Expected Annual Loss - Total": "hurricane_eal_total",
    "Hurricane - Expected Annual Loss Score": "hurricane_eal_score",
    "Hurricane - Hazard Type Risk Index Score": "hurricane_risk_score",
    "Hurricane - Hazard Type Risk Index Value": "hurricane_risk_value",
    "Hurricane - Hazard Type Risk Index Rating": "hurricane_risk_rating",
}
nri_sf_clean = nri_sf_clean.rename(columns=rename)
```

```python
# Numeric conversions
num_cols = [
    "pop_2020", "area_sqmi",
    "nri_value", "nri_score",
    "eal_total", "eal_score",
    "social_vuln_score", "resilience_score",
    "coastal_flood_eal_total", "coastal_flood_eal_score", "coastal_flood_risk_score", "coastal_flood_risk_value",
    "inland_flood_eal_total", "inland_flood_eal_score", "inland_flood_risk_score", "inland_flood_risk_value",
    "hurricane_eal_total", "hurricane_eal_score", "hurricane_risk_score", "hurricane_risk_value",
]
for c in [c for c in num_cols if c in nri_sf_clean.columns]:
    nri_sf_clean[c] = pd.to_numeric(nri_sf_clean[c], errors="coerce")
```

```python
# Checks
print("Rows (SoFlo):", len(nri_sf_clean))
print("Duplicate GEOIDs:", int(nri_sf_clean["GEOID"].duplicated().sum()))
print(nri_sf_clean[["GEOID","nri_score","eal_total"]].head())
```
